```python
#@title 4

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import urllib.request
from urllib.parse import urlparse, urlunparse
import time

url = "https://kemono.su/fantia/user/21268/post/1798991" #@param {type:"string"}

!touch links.txt
!touch parsed_url.txt

url_list = "links.txt"
parsed_url = "parsed_url.txt"

def modify_url(url):
  with open(url_list,'a',encoding="utf-8") as o:
            o.write(str(url))
  l = urlparse(url)
  with open(parsed_url,'a',encoding="utf-8") as p:
    p.write(l)
# Khởi tạo trình duyệt
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
service = Service(executable_path=r'/usr/bin/chromedriver')
driver = webdriver.Chrome(service=service,options=options)

# Mở trang web
driver.get(url)

# Cuộn trang từ đầu đến cuối để tải tất cả các hình ảnh tải lười
lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
for x in range(lenOfPage):
  lastCount = lenOfPage
#  time.sleep(3)
  screenshot = f'screenshot{x}.png'
  driver.save_screenshot(screenshot)
  lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")

'''
match=False
while(match==False):
    lastCount = lenOfPage
    time.sleep(3)
    driver.save_screenshot('screenshot.png')
    lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    if lastCount==lenOfPage:
        match=True
'''

# Tìm tất cả các thẻ img có thuộc tính 'data-src'
images = driver.find_elements_by_xpath('//img[@data-src]')

for image in images:
    # Lấy giá trị của thuộc tính 'data-src'
    url = image.get_attribute('data-src')

    # Thay đổi URL theo cấu trúc định sẵn (nếu cần)
    url = modify_url(url)

    # Tải hình ảnh về máy
    urllib.request.urlretrieve(url, '/content/downloads')

# Đóng trình duyệt
driver.quit()
```
